Cross Validation

```python
import pandas as pd
import numpy as np 

from sklearn.tree import DecisionTreeRegressor

from sklearn.model_selection import (
    cross_val_score, KFold
)
```

```python
if __name__ == "__main__":

    dataset = pd.read_csv('./data/ECommerce_consumer_behaviour.csv')

    X = dataset.drop(['reordered', 'department', 'product_name', 'days_since_prior_order'], axis=1)
    y = dataset['reordered']

    model = DecisionTreeRegressor()
    score = cross_val_score(model, X,y, cv=3, scoring='neg_mean_squared_error')
    print("="*64)
    print(score)
    print("="*64)
    print(np.abs(np.mean(score)))
    print("="*64)

    kf = KFold(n_splits=3, shuffle=True, random_state=42)
    for train, test in kf.split(dataset):
        print("="*64)
        print(train)
        print("="*64)
        print(test)
        print("="*64)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

if __name__ == "__main__":
    dataset = pd.read_csv('./data/ECommerce_consumer_behaviour.csv')

    X = dataset.drop(['reordered', 'department', 'product_name', 'days_since_prior_order'], axis=1)
    y = dataset['reordered']

    model = DecisionTreeRegressor()
    score = cross_val_score(model, X, y, cv=3, scoring='neg_mean_squared_error')
    print("="*64)
    print(score)
    print("="*64)
    print(np.abs(np.mean(score)))
    print("="*64)

    # Graficar los resultados usando Seaborn
    plt.figure(figsize=(10, 6))
    sns.set(style="whitegrid")
    sns.histplot(score, bins=10, color='#FFCE30')  # Cambiado a un solo color
    plt.title('Distribución de puntuaciones de MSE en validación cruzada')
    plt.xlabel('Puntuación de MSE')
    plt.ylabel('Frecuencia')
    plt.show()

    kf = KFold(n_splits=3, shuffle=True, random_state=42)
    for train, test in kf.split(dataset):
        print("="*64)
        print(train)
        print("="*64)
        print(test)
        print("="*64)
```
